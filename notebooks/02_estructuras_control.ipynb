```python
anio = 2001
while anio<=2012:
    print(f"Informes del año: {anio}")
    anio+=1
```

```python
# Escribe un programa que solicite al usuario ingresar un número entero
# y luego determine si es par o impar. Muestra un mensaje apropiado en función del resultado.
```

```python
numero = int(input("ingrese un numero:"))
print(f"el numero elegido fue {numero}")
validacion = numero % 2
if validacion == 0:
    print("el numero es par")
else:
    print("el numero es impar")
```

```python
#Escribe un programa que solicite al usuario ingresar un número entero
# positivo y luego calcule la suma de todos los números naturales desde 1 hasta ese número.
numero = int(input("ingrese un numero:"))
resultado = 0
if numero <=0:
    print("numero no valido")
else:
    for i in range(1,numero+1):
        resultado += i

print(f"El acumulado numerico es:{resultado}")
```
